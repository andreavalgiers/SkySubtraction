# CONTINUUM SUBTRACTION

```python
### Binning the spectra 

# Defining the parameters:

# Bin size: Should be about 2 times the width of a skyline
FWHM=10 # Average FWHM of 4 skylines
#FWHM=50.0
binwidth=2.0*FWHM
```

```python
# Defining some functions
# Defining the root-mean-square equation
#### Input is an array and output is the root-mean-square of the array
def rms(x, axis=None):
    return np.sqrt(np.mean(x**2, axis=axis))


#### Defining a function that splits a list into n parts equally.
##### Input is your list and the n parts that you want to the list to be evenly divided in. Output is the newly divided list.
def split_seq(seq, size):
        newseq = []
        splitsize = 1.0/size*len(seq)
        for i in range(size):
                newseq.append(seq[int(round(i*splitsize)):int(round((i+1)*splitsize))])
        return np.array(newseq)

    
#### Defining a function that combines the above two equation: returning the rms of each array and the equally divided lists.
def rms_sigma(wavelength,data_cube,binsize=binwidth):
    bins01=np.arange(min(wavelength), max(wavelength)+binsize, binsize)
    min_rms_data=[]
    binarr_fib=[]
    for i in range(0,len(data_cube)):
        bindata=split_seq(data_cube[i], len(bins01))
        binarr_fib.append(bindata)
        arr=[]
        for i in range(0,len(bindata)):
            arr.append(rms(bindata[i],axis=None))
            arr.sort()
        min_rms_data.append(np.mean(arr[0:3]))

    min_rms_data=np.array(min_rms_data)
    binarr_fib=np.array(binarr_fib)
    return min_rms_data,binarr_fib # Returns the min rms value and binned flux counts

# Sigma Clipping Algorithm from SciPy with rejections
    # It removes the rejected points from the array
def sigmaclip(a,std=2.0, low=10**6., high=4.):
    c = np.asarray(a).ravel()
    delta = 1
    while delta:
        c_min = c.min()
        size = c.size
        critlower = c_min - std*low
        critupper = c_min + std*high
        c = c[(c > critlower) & (c < critupper)]
        delta = size-c.size
    return c


# Sigma Clipping without rejections
    # For the rejected data points, it gives it a 0 value.
def sigmaclip_unrj(a,mean,std, low=4., high=4.):
    c = np.array(a)
    fn=[]
    critlower = mean - std*low
    critupper = mean + std*high
    
    
    for i in range(0,len(c)):
        if critlower>c[i] or c[i]>critupper:
            c[i]=0.0
    
    return np.array(c,dtype=float)
```

```python
### Running it on data

min_rms, bindata = rms_sigma(wavelength,data_cube,binsize=binwidth)

## Clipping 2*RMS away and taking the mean of each bin for each fiber.

hsig=1.0
lsig=1.0
#lsig=10**6

clp_mean_bindat=[]
for k in range(0,len(bindata)):
    arr=[]
    for i in range(0,len(bindata[0])):
        
        c=np.mean(sigmaclip(bindata[k][i],std=min_rms[k], low=lsig, high=hsig))
        arr.append(c)
    
    clp_mean_bindat.append(arr)
clp_mean_bindat=np.array(clp_mean_bindat)
clp_mean_bindat[np.isnan(clp_mean_bindat)] = 0
```

```python
### The binned wavelength list
bins01=np.arange(min(wavelength), max(wavelength)+binwidth, binwidth)
bin_wave=split_seq(wavelength, len(bins01))

median_bin_wave=[]
for i in range(0,len(bin_wave)):
    median_bin_wave.append(np.median(bin_wave[i]))
median_bin_wave=np.array(median_bin_wave)


mean_bin_wave=[]
for i in range(0,len(bin_wave)):
    mean_bin_wave.append(np.mean(bin_wave[i]))
mean_bin_wave=np.array(mean_bin_wave)
```

```python
### Plotting the Min_RMS clipped data 
n=15
plt.figure(figsize = (20,7))
x1,x2,y1,y2=8500,8800,0.0,100000

plt.plot(wavelength,data_cube[n],'k-',label='Raw Data')
plt.plot(median_bin_wave,clp_mean_bindat[n],'r-',lw=3,label='Min Binned Data')

#plt.axis([x1,x2,y1,y2])
plt.xlabel('$Wavelength \ [\AA]$')
plt.ylabel('$Flux Count$')
plt.title('Fiber '+str(n+1))
plt.legend(loc='upper left',prop={'size':14},frameon=True)
#plt.savefig('.png')

plt.show()

# Reformatting to have a multi-D array for the wavelength for each fiber
wave=[]
for i in range(0,len(data_cube)):
    wave.append(median_bin_wave)
wave=np.array(wave)

### Fitting and clipping fit from mean_clipped_data:

order=[4,5,6,7]
hsig=[8,5,4,3]
lsig=[8,5,4,3]

### This is fitting the continuum using sigma clipping and going through many iterations.

## Copy of the clipped_data

dat_sub=clp_mean_bindat[:]
wave_sub=wave[:]

for i in range(0,len(order)):
    dat_arr=[]
    wave_arr=[]
    fit_arr=[]
    for k in range(0,len(dat_sub)):
        
        # Doing the fit
        q=np.polyfit(wave_sub[k],dat_sub[k],order[i]) 
        fit=np.poly1d(q)
        fitarr=fit(wave_sub[k])
            
        # Calculate the mean of fit
        mean=np.mean(fitarr)
            
        # Calculate the standard deviation of fit
        std=np.std(fitarr)
        
        
        wave_fit=[]
        dat_fit=[]
        
        for g in range(0,len(fitarr)):
            if fitarr[g]-lsig[i]*std<dat_sub[k][g]<fitarr[g]+hsig[i]*std:
                wave_fit.append(wave_sub[k][g])
                dat_fit.append(dat_sub[k][g])
            else:
                wave_fit.append(wave_sub[k][g])
                dat_fit.append(fitarr[g])
        
        dat_arr.append(dat_fit) 
        wave_arr.append(wave_fit)
        
        fit_arr.append(fitarr)
    dat_sub=dat_arr[:] # The Data that has not been clipped
    #dat_sub=fit_arr[:] # The fit of the unclipped data
    wave_sub=wave_arr[:]  
    


### Plotting Original, Binned, and Clipped_To_Fit Data:
plt.figure(figsize = (20,7))
x1,x2,y1,y2=8500,8800,0.0,100000

plt.plot(wavelength,data_cube[n],'k-',label='Raw Data')
plt.plot(median_bin_wave,clp_mean_bindat[n],'r-',lw=3,label='Binned Data')
plt.plot(wave_sub[n],dat_sub[n],'bo',lw=3,label='Data Left Over from Clipping')

#plt.axis([x1,x2,y1,y2])
plt.xlabel('$Wavelength \ [\AA]$')
plt.ylabel('$Flux Count$')
plt.title('Fiber '+str(n+1))
plt.legend(loc='upper left',prop={'size':14},frameon=True)
#plt.savefig('.png')

plt.show()


### Interpolating the fitted data

dat_interp=[]

for i in range(0,len(dat_sub)):
    s = smoothn.smoothn(np.array(dat_sub[i]),s=2) # Spline Smooth
    f = scipy.interpolate.interp1d(wave_sub[i] ,s[0],kind='cubic',bounds_error=False,fill_value=s[0][-1]) # Interpolating the Spline
    dat_interp.append(f(wavelength))
        
dat_interp=np.array(dat_interp)


### Plotting Original, Binned, and Interp Data:
plt.figure(figsize = (20,7))
x1,x2,y1,y2=6500,7500,10000,25000

plt.plot(wavelength,data_cube[n],'k-',label='Raw Data')
#plt.plot(median_bin_wave,clp_mean_bindat[n],'r-',lw=3,label='Binned Data')
plt.plot(wave_sub[n],dat_sub[n],'r-',lw=3,label='Data Left Over from Clipping')
plt.plot(wavelength,dat_interp[n],'b-',lw=3,label='Spline Fit')

#plt.axis([x1,x2,y1,y2])
plt.xlabel('$Wavelength \ [\AA]$')
plt.ylabel('$Flux Count$')
plt.title('Fiber '+str(n+1))
plt.legend(loc='upper left',prop={'size':14},frameon=True)
#plt.savefig('.png')

plt.show()
```
